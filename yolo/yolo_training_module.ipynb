```python
task = Task.init(
    project_name="training_yolov8_in_kaggle",
    task_name="training_yolov8_in_kaggle"
)

model_variant = "yolo8m"
task.set_parameter("model_variant", model_variant)

model = YOLO('yolov8m.pt')
```

```python
path_to_parameters = '/kaggle/input/object-detection-swans/DATA_FOR_YOLO'

args = dict(
   imgsz=1280,
   epochs=100,
   batch=8,
   data=f'{path_to_parameters}/data.yaml',
   name=f'{model_variant}_custom'
)

task.connect(args)
```
